La libreria `missingno` di Python Ã¨ un'ottima soluzione per visualizzare e analizzare i dati mancanti (missing data) in un DataFrame. Permette di esplorare facilmente la distribuzione e il pattern dei valori mancanti in un dataset, facilitando il processo di pulizia dei dati.

````python
import missingno as msno
import pandas as pd
````

````python
df = pd.read_csv("animal-data-1.csv")

print(df.info())
````

### Funzioni principali

1. **Visualizzare la distribuzione dei dati mancanti**
    - `missingno.matrix()`: Visualizza una matrice che mostra la posizione dei valori mancanti. I valori presenti sono mostrati in bianco, mentre i valori mancanti sono in nero. Questo aiuta a capire se ci sono schemi nei dati mancanti.

````python
msno.matrix(df)
````

2. **Heatmap dei dati mancanti**
    - `missingno.heatmap()`: Crea una heatmap che mostra le correlazioni tra le colonne basate sulla presenza o assenza di dati. Se due colonne hanno valori mancanti nello stesso record, sono correlate, e la heatmap lo evidenzia.

````python
msno.heatmap(df)
````

3. **Visualizzare la distribuzione delle assenze**
    - `missingno.dendrogram()`: Visualizza un dendrogramma per raggruppare le colonne in base alla correlazione tra i valori mancanti.
    ```

````python
msno.dendrogram(df)
````

4. **Bar plot della percentuale di valori mancanti**
    - `missingno.bar()`: Mostra un grafico a barre della percentuale di dati mancanti in ogni colonna.

````python
msno.bar(df)
````

5. **Distribuzione dei dati mancanti per singolo valore**
    - `missingno.nullity_filter()`: Mostra un filtro che ti permette di visualizzare quali colonne o righe hanno la percentuale piÃ¹ alta di dati mancanti.

````python
msno.nullity_filter(df)
````
